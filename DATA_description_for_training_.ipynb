```python
# Create a 'year' column from 'Actual Opening date'
df['year'] = pd.to_numeric(df['Actual Opening date'], errors='coerce')
# Fill any NaN values that resulted from coercion or original NaNs, then convert to integer
df['year'] = df['year'].fillna(0).astype(int)

df['period'] = pd.cut(
    df['year'],
    bins=[2004, 2010, 2015, 2020, 2025],
    labels=['2005–2010', '2011–2015', '2016–2020', '2021–2025']
)

table_41_period = (
    df['period']
    .value_counts()
    .rename_axis('Category')
    .reset_index(name='N')
)

table_41_period['%'] = round(table_41_period['N'] / len(df) * 100, 1)
table_41_period
```

```python
def freq_table(column):
    t = (
        df[column]
        .value_counts()
        .rename_axis('Category')
        .reset_index(name='N')
    )
    t['%'] = round(t['N'] / len(df) * 100, 1)
    return t

# Display all column names to help identify the correct ones
print("Available columns in DataFrame df:")
print(df.columns.tolist())

# table_41_contract = freq_table('contract_type')
# table_41_procurement = freq_table('procurement_method')
# table_41_financing = freq_table('financing_structure')
```

```python
table_42_region = freq_table('Region')
table_42_region
```

```python
df_clean = df.rename(columns={
    'Region': 'region',
    'Industry( Sector)': 'sector',
    'Revenue Model Type ': 'financing_structure',
    'Project Financing ': 'project_financing',
    'PPP Procuerment Process Level': 'ppp_procurement_quality',
    'Risk Allocations ': 'risk_allocation',
    'Smart Innovation ': 'smart_innovation',
    'Financial Viability ': 'financial_viability',
    'Evaluation ( 1 for success as the project in operation or 0)': 'overall_success',
    'year': 'year',
    'period': 'period',
    'Governance 1-100': 'governance_index',
    ' Regulatory Quality Assess and Institutional Quality) ': 'regulatory_quality',
    'Control of Corruption (accurancy of procuerment process) ': 'control_of_corruption',
    'Renegotiation': 'renegotiation'
})
```

```python
df_clean['time_success'] = (df_clean['Lifecycle Cost '] == 1).astype(int)
df_clean['cost_success'] = (df_clean['Lifecycle Cost '] == 1).astype(int)
df_clean['quality_success'] = (df_clean['Partnership and collabration '] == 1).astype(int)
```

```python
df_clean['institutional_contract_enforcement'] = (
    pd.to_numeric(df_clean['governance_index'], errors='coerce') / 100
)
```

```python
df_clean.shape

```

```python
governance_vars = [
    'regulatory_quality',
    'ppp_procurement_quality',
    'risk_allocation'
]

gov_stats = []

for var in governance_vars:
    gov_stats.append({
        'Variable': var,
        'Mean': round(df_clean[var].mean(), 2),
        'SD': round(df_clean[var].std(), 2),
        'Min': round(df_clean[var].min(), 2),
        'Max': round(df_clean[var].max(), 2)
    })

table_44 = pd.DataFrame(gov_stats)
table_44
```
