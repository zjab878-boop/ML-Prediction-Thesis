```python
import pandas as pd
import numpy as np

# -----------------------------
# 0) (You already loaded df, but keep this line if you restart kernel)
# df = pd.read_csv("ZizfonJdatabase.csv", encoding="latin1")
# -----------------------------

# Clean column names (removes hidden spaces)
df.columns = df.columns.str.strip()

# -----------------------------
# 1) Outcome + conditions
# -----------------------------
outcome_col = "Evaluation"

condition_cols = [
    "Renegotiation",
    "RevenueModel_type",
    "VoiceOfAccountability_StakeholderEngagement",
    "Macro_Enviromentimpact",
    "Governance",
    "Regulority_Qualityassess",
    "ControlofCorruption",
    "BiddingTeam",
    "PPP_Procurement",
    "Financial_capacity",
    "Project_Financing",
    "Financial_Viability",
    "Partnership_collabration",
    "Lifecycle_cost",
    "Political_Risk",
    "Sustainability",
    "Risk_allocation",
    "Contrcatural_Duties",
    "Cost_efficiency",
    "Environmental_Impact",
    "Insitutional Contrcat Enforcement",
    "StakeholderSatisfaction",
    "Outcome_specification",
    "Service_Quality",
    "Time",
    "Cost",
    "Quality",
    "Smart_Innovation",
    "concession_period"
]

# -----------------------------
# 2) Direct fuzzy calibration helpers
# -----------------------------
def guess_anchors(series: pd.Series):
    s = pd.to_numeric(series, errors="coerce")
    smin, smax = np.nanmin(s), np.nanmax(s)

    # Likert detection
    if smin >= 1 and smax <= 5:
        return 1, 3, 5
    if smin >= 1 and smax <= 7:
        return 1, 4, 7
    if smin >= 1 and smax <= 9:
        return 1, 5, 9

    # Fallback: robust quantiles
    q10, q50, q90 = np.nanquantile(s, [0.10, 0.50, 0.90])
    return float(q10), float(q50), float(q90)

def calibrate_fuzzy_direct(x, low, crossover, high):
    x = pd.to_numeric(x, errors="coerce")
    # logit(0.95) ~ 2.944
    c = np.log(0.95 / 0.05)
    denom = (high - crossover)
    if denom == 0:
        denom = 1e-9
    k = c / denom
    f = 1 / (1 + np.exp(-k * (x - crossover)))
    return np.clip(f, 0, 1)

# -----------------------------
# 3) Reverse-direction variables (high = bad for success)
#    Edit this set ONLY if your coding is already "higher = better"
# -----------------------------
reverse_vars = {
    "Political_Risk",
    "Renegotiation",
    "Lifecycle_cost",
    "Environmental_Impact",
    "Macro_Enviromentimpact"
}

# -----------------------------
# 4) Build the final fuzzy QCA dataset
# -----------------------------
# Keep identifiers separately (optional, NOT used in QCA minimization)
id_cols = [c for c in ["YEAR", "Sector", "Statu_ID"] if c in df.columns]

qca = pd.DataFrame()

# Outcome
low, cross, high = guess_anchors(df[outcome_col])
qca["Y"] = calibrate_fuzzy_direct(df[outcome_col], low, cross, high)

# Conditions
missing = []
for col in condition_cols:
    if col not in df.columns:
        missing.append(col)
        continue

    low, cross, high = guess_anchors(df[col])
    fz = calibrate_fuzzy_direct(df[col], low, cross, high)

    if col in reverse_vars:
        fz = 1 - fz

    qca[f"f_{col}"] = fz

if missing:
    print(" These condition columns were not found in your file (skipped):")
    print(missing)

# Drop rows with any missing fuzzy values
qca_clean = qca.dropna().copy()

print(" Final fuzzy dataset created")
print("Raw rows:", df.shape[0], "â†’ QCA rows after dropna:", qca_clean.shape[0])
print("Columns:", qca_clean.shape[1])

# Save
qca_clean.to_csv("PPP_QCA_FINAL_FUZZY.csv", index=False)
print(" Saved: PPP_QCA_FINAL_FUZZY.csv")

# Quick peek
qca_clean.head()
```

```python
def necessity_metrics(X, Y):
    X = np.asarray(X, dtype=float)
    Y = np.asarray(Y, dtype=float)

    # Consistency of necessity: sum(min(X,Y)) / sum(Y)
    cons = np.sum(np.minimum(X, Y)) / np.sum(Y) if np.sum(Y) != 0 else np.nan

    # Coverage of necessity: sum(min(X,Y)) / sum(X)
    cov = np.sum(np.minimum(X, Y)) / np.sum(X) if np.sum(X) != 0 else np.nan

    return cons, cov

Y = qca_clean["Y"].values
cond_cols = [c for c in qca_clean.columns if c.startswith("f_")]

rows = []
for c in cond_cols:
    cons, cov = necessity_metrics(qca_clean[c].values, Y)
    rows.append((c, cons, cov))

nec = pd.DataFrame(rows, columns=["Condition", "Necessity_Consistency", "Necessity_Coverage"])
nec = nec.sort_values("Necessity_Consistency", ascending=False)

nec.head(15)
```

```python
from sklearn.model_selection import train_test_split

# Define X and y
feature_cols = [
    c for c in df.columns 
    if c not in [
        "Country_ID", "PROJECT_Name", "Project_Statu", "Statu_ID",
        "Sector", "YEAR", "Evaluation", "success_score", "Y_strict"
    ]
]

X = df[feature_cols].apply(pd.to_numeric, errors="coerce")
y = df["Y_strict"]

# Drop missing
data = pd.concat([X, y], axis=1).dropna()
X = data[feature_cols]
y = data["Y_strict"]

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    random_state=42,
    stratify=y
)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
print("Train class balance:\n", y_train.value_counts())
print("Test class balance:\n", y_test.value_counts())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report, confusion_matrix

# Logistic Regression with class weighting
logit = LogisticRegression(
    max_iter=1000,
    class_weight="balanced",
    solver="lbfgs"
)

logit.fit(X_train, y_train)

# Predictions
y_pred = logit.predict(X_test)
y_proba = logit.predict_proba(X_test)[:, 1]

# Evaluation
print("LOGISTIC REGRESSION â€“ TEST RESULTS\n")
print("Accuracy:", round(accuracy_score(y_test, y_pred), 3))
print("ROC-AUC :", round(roc_auc_score(y_test, y_proba), 3))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# Generate sample data
X, y = make_classification(n_samples=1000, n_classes=2, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a logistic regression model
logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)
y_prob_logistic = logistic_model.predict_proba(X_test)[:, 1]  # Get probabilities for positive class

# Train a random forest model
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
y_prob_rf = rf_model.predict_proba(X_test)[:, 1]  # Get probabilities for positive class

# Calculate ROC curves
fpr_log, tpr_log, _ = roc_curve(y_test, y_prob_logistic)
roc_auc_log = auc(fpr_log, tpr_log)

fpr_rf, tpr_rf, _ = roc_curve(y_test, y_prob_rf)
roc_auc_rf = auc(fpr_rf, tpr_rf)

# Plot ROC curves
plt.figure()
plt.plot(fpr_log, tpr_log, label=f'Logistic Regression (AUC = {roc_auc_log:.2f})')
plt.plot(fpr_rf, tpr_rf, label=f'Random Forest (AUC = {roc_auc_rf:.2f})')
plt.plot([0, 1], [0, 1], linestyle='--', color='grey', label='Random classifier')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curves')
plt.legend()
plt.show()
```
