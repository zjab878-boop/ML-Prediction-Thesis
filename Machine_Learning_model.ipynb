```python
import numpy as np
import pandas as pd

Y_num = Y.apply(pd.to_numeric, errors="coerce")

# Drop rows with missing outcome values (recommended)
mask = Y_num.notna().all(axis=1)
X = X.loc[mask].copy()
Y_num = Y_num.loc[mask].copy()

Y_bin = pd.DataFrame(index=Y_num.index)

for col in Y_num.columns:
    vals = set(Y_num[col].unique())

    # already binary
    if vals.issubset({0, 1}):
        Y_bin[col] = Y_num[col].astype(int)

    # fuzzy 0–1
    elif Y_num[col].min() >= 0 and Y_num[col].max() <= 1:
        Y_bin[col] = (Y_num[col] >= 0.5).astype(int)

    # ordinal / other numeric → median split
    else:
        thr = Y_num[col].median()
        Y_bin[col] = (Y_num[col] >= thr).astype(int)

Y_bin.apply(pd.Series.value_counts)
```

```python
import pandas as pd
import numpy as np
import difflib
import traceback

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

try:
    print("STEP 1: Load CSV")
    df = pd.read_csv("/content/PPP_ML_clean.csv")
    print("  df:", df.shape)

    print("STEP 2: Clean column names")
    df.columns = (
        df.columns.astype(str)
        .str.strip()
        .str.replace(" ", "_")
        .str.replace("-", "_")
    )
    print("  first 15 cols:", df.columns[:15].tolist())

    print("STEP 3: Resolve outcome columns")
    cols = df.columns.tolist()
    cols_lower = {c.lower(): c for c in cols}

    def resolve_col(name: str):
        key = name.lower()
        if key in cols_lower:
            return cols_lower[key]
        matches = difflib.get_close_matches(key, cols_lower.keys(), n=3, cutoff=0.75)
        return cols_lower[matches[0]] if matches else None

    requested_outcomes = [
        "Time", "Cost", "Quality", "Financial_Viability",
        "Institutional_Contract_Enforcement",
        "StakeholderSatisfaction", "Smart_Innovation", "Service_Quality"
    ]

    outcome_cols = [resolve_col(c) for c in requested_outcomes]
    missing = [req for req, got in zip(requested_outcomes, outcome_cols) if got is None]
    if missing:
        print(" Missing outcome columns:", missing)
        print(" Available columns sample:", df.columns[:80].tolist())
        raise KeyError("Missing outcome columns (see above).")

    print("  outcomes resolved:", outcome_cols)

    print("STEP 4: Prepare Y (binarise) and align data")
    Y_raw = df[outcome_cols].apply(pd.to_numeric, errors="coerce")

    # Find rows where all raw outcome values are non-null
    Y_valid_rows_mask = Y_raw.notna().all(axis=1)

    # Filter df to only include rows with valid outcomes
    df_filtered = df.loc[Y_valid_rows_mask].copy()
    Y_raw_filtered = Y_raw.loc[Y_valid_rows_mask].copy() # Ensure Y_raw_filtered is also masked

    # Binarize each outcome based on its distribution
    Y_bin = pd.DataFrame(index=Y_raw_filtered.index)
    for col in Y_raw_filtered.columns:
        unique_vals = Y_raw_filtered[col].unique()
        if len(unique_vals) <= 1: # If all values are the same or only NaN, it's constant
            Y_bin[col] = 0 # Assigning 0, will be filtered out by nunique check
        elif set(unique_vals).issubset({0, 1}): # Already binary (0 or 1)
            Y_bin[col] = Y_raw_filtered[col].astype(int)
        elif Y_raw_filtered[col].min() >= 0 and Y_raw_filtered[col].max() <= 1: # Fuzzy 0-1 range
            Y_bin[col] = (Y_raw_filtered[col] >= 0.5).astype(int)
        else: # Ordinal or other numeric, use median split
            median_val = Y_raw_filtered[col].median()
            Y_bin[col] = (Y_raw_filtered[col] >= median_val).astype(int)

    print(f"  Y_bin shape after binarization: {Y_bin.shape}")

    print("STEP 5: Drop constant outcomes")
    valid_outcomes = [c for c in Y_bin.columns if Y_bin[c].nunique() == 2]
    Y_bin2 = Y_bin[valid_outcomes].copy()
    print("  valid outcomes:", list(Y_bin2.columns))

    print("STEP 6: Drop extremely imbalanced outcomes safely")
    for col in list(Y_bin2.columns):
        vc = Y_bin2[col].value_counts(dropna=True)
        if vc.empty:
            print(f"   Dropping empty outcome: {col}")
            Y_bin2 = Y_bin2.drop(columns=[col])
            continue
        if vc.min() < 5: # Keep the min 5 rule, but now more columns should pass due to better binarization
            print(f"   Dropping imbalanced outcome: {col} -> {vc.to_dict()}")
            Y_bin2 = Y_bin2.drop(columns=[col])

    if Y_bin2.shape[1] == 0:
        raise ValueError("After dropping constant/imbalanced outcomes, Y has 0 columns. Lower the imbalance rule or adjust threshold.")

    print("  ✅ Outcomes used:", list(Y_bin2.columns), "| Y shape:", Y_bin2.shape)

    print("STEP 7: Build X (numeric conversion)")
    X_raw = df_filtered.drop(columns=outcome_cols, errors="ignore").copy() # Use df_filtered here
    X_num = X_raw.apply(pd.to_numeric, errors="coerce")
    X_num = X_num.dropna(axis=1, how="all")

    low_var_cols = [c for c in X_num.columns if X_num[c].nunique(dropna=True) <= 1]
    if low_var_cols:
        print("  dropping low-variance cols:", low_var_cols)
        X_num = X_num.drop(columns=low_var_cols)

    X = X_num.copy()
    print("   X shape:", X.shape)

    if X.shape[1] == 0:
        raise ValueError("X has 0 columns after numeric conversion. Your predictors may be categorical text; use one-hot encoding.")

    print("STEP 8: Align X and Y")
    # X_aligned and Y_aligned are already implicitly aligned because they both derived from df_filtered
    X_aligned = X # X is already X_num from df_filtered
    Y_aligned = Y_bin2 # Y_bin2 is already derived from Y_raw_filtered

    print("   aligned:", X_aligned.shape, Y_aligned.shape)

    print("STEP 9: Split (safe stratify; fallback if it fails)")
    success_count = Y_aligned.sum(axis=1)

    strata = None
    try:
        strata = pd.qcut(success_count, q=4, labels=False, duplicates="drop")
        if strata.value_counts().min() < 2:
            strata = None
    except Exception: # Catch any error during qcut if bins are not feasible
        strata = None

    if strata is None:
        print("   Stratify not used (fallback).")
        X_train, X_test, Y_train, Y_test = train_test_split(
            X_aligned, Y_aligned, test_size=0.30, random_state=42
        )
    else:
        # Ensure strata is also filtered to match X_aligned/Y_aligned
        strata_final = strata.loc[Y_aligned.index] # Align strata to Y_aligned's index
        X_train, X_test, Y_train, Y_test = train_test_split(
            X_aligned, Y_aligned, test_size=0.30, random_state=42, stratify=strata_final
        )

    print("   TRAIN:", X_train.shape, Y_train.shape)
    print("   TEST :", X_test.shape, Y_test.shape)

    print("STEP 10: Train Random Forest (multi-output)")
    rf_model = Pipeline(steps=[
        ("imputer", SimpleImputer(strategy="median")),
        ("clf", MultiOutputClassifier(
            RandomForestClassifier(
                n_estimators=400,
                random_state=42,
                class_weight="balanced_subsample",
                n_jobs=-1
            )
        ))
    ])
    rf_model.fit(X_train, Y_train)
    print("  ✅ RF trained")

    print("STEP 11: Feature importance")
    rf_estimators = rf_model.named_steps["clf"].estimators_
    feature_names = X_train.columns.tolist()
    outcomes = Y_train.columns.tolist()

    fi = pd.DataFrame(
        {out: est.feature_importances_ for out, est in zip(outcomes, rf_estimators)},
        index=feature_names
    )

    fi["Global_Mean"] = fi.mean(axis=1)
    fi_global = fi.sort_values("Global_Mean", ascending=False).head(15)

    print("\n🏁 Top 15 Global RF Features:")
    print(fi_global.to_string())

    print("\n🏁 Top 10 RF Features by Outcome:")
    top_k = 10
    for out in outcomes:
        print("\n" + "="*60)
        print("Outcome:", out)
        print(fi[out].sort_values(ascending=False).head(top_k).to_string())

except Exception as e:
    print("\n FAILED. Exact error:")
    print(type(e).__name__ + ":", str(e))
    traceback.print_exc()
```

```python
import pandas as pd
import numpy as np
import difflib
import traceback

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

try:
    print("STEP 1: Load CSV")
    df = pd.read_csv("/content/PPP_ML_clean.csv") # Changed to PPP_ML_clean.csv
    print("  df:", df.shape)

    print("STEP 2: Clean column names")
    df.columns = (
        df.columns.astype(str)
        .str.strip()
        .str.replace(" ", "_")
        .str.replace("-", "_")
    )
    print("  first 15 cols:", df.columns[:15].tolist())

    print("STEP 3: Resolve outcome columns")
    cols = df.columns.tolist()
    cols_lower = {c.lower(): c for c in cols}

    def resolve_col(name: str):
        key = name.lower()
        if key in cols_lower:
            return cols_lower[key]
        matches = difflib.get_close_matches(key, cols_lower.keys(), n=3, cutoff=0.75)
        return cols_lower[matches[0]] if matches else None

    requested_outcomes = [
        "Time", "Cost", "Quality", "Financial_Viability",
        "Institutional_Contract_Enforcement",
        "StakeholderSatisfaction", "Smart_Innovation", "Service_Quality"
    ]

    outcome_cols = [resolve_col(c) for c in requested_outcomes]
    missing = [req for req, got in zip(requested_outcomes, outcome_cols) if got is None]
    if missing:
        print(" Missing outcome columns:", missing)
        print(" Available columns sample:", df.columns[:80].tolist())
        raise KeyError("Missing outcome columns (see above).")

    print("  outcomes resolved:", outcome_cols)

    print("STEP 4: Prepare Y (binarise) and align data")
    Y_raw = df[outcome_cols].apply(pd.to_numeric, errors="coerce")

    # Find rows where all raw outcome values are non-null
    Y_valid_rows_mask = Y_raw.notna().all(axis=1)

    # Filter df to only include rows with valid outcomes
    df_filtered = df.loc[Y_valid_rows_mask].copy()
    Y_raw_filtered = Y_raw.loc[Y_valid_rows_mask].copy() # Ensure Y_raw_filtered is also masked

    # Binarize each outcome based on its distribution
    Y_bin = pd.DataFrame(index=Y_raw_filtered.index)
    for col in Y_raw_filtered.columns:
        unique_vals = Y_raw_filtered[col].unique()
        if len(unique_vals) <= 1: # If all values are the same or only NaN, it's constant
            Y_bin[col] = 0 # Assigning 0, will be filtered out by nunique check
        elif set(unique_vals).issubset({0, 1}): # Already binary (0 or 1)
            Y_bin[col] = Y_raw_filtered[col].astype(int)
        elif Y_raw_filtered[col].min() >= 0 and Y_raw_filtered[col].max() <= 1: # Fuzzy 0-1 range
            Y_bin[col] = (Y_raw_filtered[col] >= 0.5).astype(int)
        else: # Ordinal or other numeric, use median split
            median_val = Y_raw_filtered[col].median()
            Y_bin[col] = (Y_raw_filtered[col] >= median_val).astype(int)

    print(f"  Y_bin shape after binarization: {Y_bin.shape}")

    print("STEP 5: Drop constant outcomes")
    valid_outcomes = [c for c in Y_bin.columns if Y_bin[c].nunique() == 2]
    Y_bin2 = Y_bin[valid_outcomes].copy()
    print("  valid outcomes:", list(Y_bin2.columns))

    print("STEP 6: Drop extremely imbalanced outcomes safely")
    for col in list(Y_bin2.columns):
        vc = Y_bin2[col].value_counts(dropna=True)
        if vc.empty:
            print(f"   Dropping empty outcome: {col}")
            Y_bin2 = Y_bin2.drop(columns=[col])
            continue
        if vc.min() < 5: # Keep the min 5 rule, but now more columns should pass due to better binarization
            print(f"   Dropping imbalanced outcome: {col} -> {vc.to_dict()}")
            Y_bin2 = Y_bin2.drop(columns=[col])

    if Y_bin2.shape[1] == 0:
        raise ValueError("After dropping constant/imbalanced outcomes, Y has 0 columns. Lower the imbalance rule or adjust threshold.")

    print("  ✅ Outcomes used:", list(Y_bin2.columns), "| Y shape:", Y_bin2.shape)

    print("STEP 7: Build X (numeric conversion)")
    X_raw = df_filtered.drop(columns=outcome_cols, errors="ignore").copy() # Use df_filtered here
    X_num = X_raw.apply(pd.to_numeric, errors="coerce")
    X_num = X_num.dropna(axis=1, how="all")

    low_var_cols = [c for c in X_num.columns if X_num[c].nunique(dropna=True) <= 1]
    if low_var_cols:
        print("  dropping low-variance cols:", low_var_cols)
        X_num = X_num.drop(columns=low_var_cols)

    X = X_num.copy()
    print("   X shape:", X.shape)

    if X.shape[1] == 0:
        raise ValueError("X has 0 columns after numeric conversion. Your predictors may be categorical text; use one-hot encoding.")

    print("STEP 8: Align X and Y")
    # X_aligned and Y_aligned are already implicitly aligned because they both derived from df_filtered
    X_aligned = X # X is already X_num from df_filtered
    Y_aligned = Y_bin2 # Y_bin2 is already derived from Y_raw_filtered

    print("   aligned:", X_aligned.shape, Y_aligned.shape)

    print("STEP 9: Split (safe stratify; fallback if it fails)")
    success_count = Y_aligned.sum(axis=1)

    strata = None
    try:
        strata = pd.qcut(success_count, q=4, labels=False, duplicates="drop")
        if strata.value_counts().min() < 2:
            strata = None
    except Exception: # Catch any error during qcut if bins are not feasible
        strata = None

    if strata is None:
        print("   Stratify not used (fallback).")
        X_train, X_test, Y_train, Y_test = train_test_split(
            X_aligned, Y_aligned, test_size=0.30, random_state=42
        )
    else:
        # Ensure strata is also filtered to match X_aligned/Y_aligned
        strata_final = strata.loc[Y_aligned.index] # Align strata to Y_aligned's index
        X_train, X_test, Y_train, Y_test = train_test_split(
            X_aligned, Y_aligned, test_size=0.30, random_state=42, stratify=strata_final
        )

    print("   TRAIN:", X_train.shape, Y_train.shape)
    print("   TEST :", X_test.shape, Y_test.shape)

    print("STEP 10: Train Random Forest (multi-output)")
    rf_model = Pipeline(steps=[
        ("imputer", SimpleImputer(strategy="median")),
        ("clf", MultiOutputClassifier(
            RandomForestClassifier(
                n_estimators=400,
                random_state=42,
                class_weight="balanced_subsample",
                n_jobs=-1
            )
        ))
    ])
    rf_model.fit(X_train, Y_train)
    print("  ✅ RF trained")

    print("STEP 11: Feature importance")
    rf_estimators = rf_model.named_steps["clf"].estimators_
    feature_names = X_train.columns.tolist()
    outcomes = Y_train.columns.tolist()

    fi = pd.DataFrame(
        {out: est.feature_importances_ for out, est in zip(outcomes, rf_estimators)},
        index=feature_names
    )

    fi["Global_Mean"] = fi.mean(axis=1)
    fi_global = fi.sort_values("Global_Mean", ascending=False).head(15)

    print("\n🏁 Top 15 Global RF Features:")
    print(fi_global.to_string())

    print("\n🏁 Top 10 RF Features by Outcome:")
    top_k = 10
    for out in outcomes:
        print("\n" + "="*60)
        print("Outcome:", out)
        print(fi[out].sort_values(ascending=False).head(top_k).to_string())

except Exception as e:
    print("\n FAILED. Exact error:")
    print(type(e).__name__ + ":", str(e))
    traceback.print_exc()
```

```python
import pandas as pd
import numpy as np
import difflib
import traceback

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

try:
    print("STEP 1: Load CSV")
    df = pd.read_csv("/content/PPP_ML_File.(1).csv") # Changed to PPP_ML_File.(1).csv as requested
    print("  df:", df.shape)

    print("STEP 2: Clean column names")
    df.columns = (
        df.columns.astype(str)
        .str.strip()
        .str.replace(" ", "_")
        .str.replace("-", "_")
    )
    print("  first 15 cols:", df.columns[:15].tolist())

    print("STEP 3: Resolve outcome columns")
    cols = df.columns.tolist()
    cols_lower = {c.lower(): c for c in cols}

    def resolve_col(name: str):
        key = name.lower()
        if key in cols_lower:
            return cols_lower[key]
        matches = difflib.get_close_matches(key, cols_lower.keys(), n=3, cutoff=0.75)
        return cols_lower[matches[0]] if matches else None

    requested_outcomes = [
        "Time", "Cost", "Quality", "Financial_Viability",
        "Institutional_Contract_Enforcement",
        "StakeholderSatisfaction", "Smart_Innovation", "Service_Quality"
    ]

    outcome_cols = [resolve_col(c) for c in requested_outcomes]
    missing = [req for req, got in zip(requested_outcomes, outcome_cols) if got is None]
    if missing:
        print(" Missing outcome columns:", missing)
        print(" Available columns sample:", df.columns[:80].tolist())
        raise KeyError("Missing outcome columns (see above).")

    print("  outcomes resolved:", outcome_cols)

    print("STEP 4: Prepare Y (binarise) and align data")
    Y_raw = df[outcome_cols].apply(pd.to_numeric, errors="coerce")

    # Find rows where all raw outcome values are non-null
    Y_valid_rows_mask = Y_raw.notna().all(axis=1)

    # Filter df to only include rows with valid outcomes
    df_filtered = df.loc[Y_valid_rows_mask].copy()
    Y_raw_filtered = Y_raw.loc[Y_valid_rows_mask].copy() # Ensure Y_raw_filtered is also masked

    # Binarize each outcome based on its distribution
    Y_bin = pd.DataFrame(index=Y_raw_filtered.index)
    for col in Y_raw_filtered.columns:
        unique_vals = Y_raw_filtered[col].unique()
        if len(unique_vals) <= 1: # If all values are the same or only NaN, it's constant
            Y_bin[col] = 0 # Assigning 0, will be filtered out by nunique check
        elif set(unique_vals).issubset({0, 1}): # Already binary (0 or 1)
            Y_bin[col] = Y_raw_filtered[col].astype(int)
        elif Y_raw_filtered[col].min() >= 0 and Y_raw_filtered[col].max() <= 1: # Fuzzy 0-1 range
            Y_bin[col] = (Y_raw_filtered[col] >= 0.5).astype(int)
        else: # Ordinal or other numeric, use median split
            median_val = Y_raw_filtered[col].median()
            Y_bin[col] = (Y_raw_filtered[col] >= median_val).astype(int)

    print(f"  Y_bin shape after binarization: {Y_bin.shape}")

    print("STEP 5: Drop constant outcomes")
    valid_outcomes = [c for c in Y_bin.columns if Y_bin[c].nunique() == 2]
    Y_bin2 = Y_bin[valid_outcomes].copy()
    print("  valid outcomes:", list(Y_bin2.columns))

    print("STEP 6: Drop extremely imbalanced outcomes safely")
    for col in list(Y_bin2.columns):
        vc = Y_bin2[col].value_counts(dropna=True)
        if vc.empty:
            print(f"   Dropping empty outcome: {col}")
            Y_bin2 = Y_bin2.drop(columns=[col])
            continue
        if vc.min() < 5: # Keep the min 5 rule, but now more columns should pass due to better binarization
            print(f"   Dropping imbalanced outcome: {col} -> {vc.to_dict()}")
            Y_bin2 = Y_bin2.drop(columns=[col])

    if Y_bin2.shape[1] == 0:
        raise ValueError("After dropping constant/imbalanced outcomes, Y has 0 columns. Lower the imbalance rule or adjust threshold.")

    print("  ✅ Outcomes used:", list(Y_bin2.columns), "| Y shape:", Y_bin2.shape)

    print("STEP 7: Build X (numeric conversion)")
    X_raw = df_filtered.drop(columns=outcome_cols, errors="ignore").copy() # Use df_filtered here
    X_num = X_raw.apply(pd.to_numeric, errors="coerce")
    X_num = X_num.dropna(axis=1, how="all")

    low_var_cols = [c for c in X_num.columns if X_num[c].nunique(dropna=True) <= 1]
    if low_var_cols:
        print("  dropping low-variance cols:", low_var_cols)
        X_num = X_num.drop(columns=low_var_cols)

    X = X_num.copy()
    print("   X shape:", X.shape)

    if X.shape[1] == 0:
        raise ValueError("X has 0 columns after numeric conversion. Your predictors may be categorical text; use one-hot encoding.")

    print("STEP 8: Align X and Y")
    # X_aligned and Y_aligned are already implicitly aligned because they both derived from df_filtered
    X_aligned = X # X is already X_num from df_filtered
    Y_aligned = Y_bin2 # Y_bin2 is already derived from Y_raw_filtered

    print("   aligned:", X_aligned.shape, Y_aligned.shape)

    print("STEP 9: Split (safe stratify; fallback if it fails)")
    success_count = Y_aligned.sum(axis=1)

    strata = None
    try:
        strata = pd.qcut(success_count, q=4, labels=False, duplicates="drop")
        if strata.value_counts().min() < 2:
            strata = None
    except Exception: # Catch any error during qcut if bins are not feasible
        strata = None

    if strata is None:
        print("   Stratify not used (fallback).")
        X_train, X_test, Y_train, Y_test = train_test_split(
            X_aligned, Y_aligned, test_size=0.30, random_state=42
        )
    else:
        # Ensure strata is also filtered to match X_aligned/Y_aligned
        strata_final = strata.loc[Y_aligned.index] # Align strata to Y_aligned's index
        X_train, X_test, Y_train, Y_test = train_test_split(
            X_aligned, Y_aligned, test_size=0.30, random_state=42, stratify=strata_final
        )

    print("   TRAIN:", X_train.shape, Y_train.shape)
    print("   TEST :", X_test.shape, Y_test.shape)

    print("STEP 10: Train Random Forest (multi-output)")
    rf_model = Pipeline(steps=[
        ("imputer", SimpleImputer(strategy="median")),
        ("clf", MultiOutputClassifier(
            RandomForestClassifier(
                n_estimators=400,
                random_state=42,
                class_weight="balanced_subsample",
                n_jobs=-1
            )
        ))
    ])
    rf_model.fit(X_train, Y_train)
    print("  ✅ RF trained")

    print("STEP 11: Feature importance")
    rf_estimators = rf_model.named_steps["clf"].estimators_
    feature_names = X_train.columns.tolist()
    outcomes = Y_train.columns.tolist()

    fi = pd.DataFrame(
        {out: est.feature_importances_ for out, est in zip(outcomes, rf_estimators)},
        index=feature_names
    )

    fi["Global_Mean"] = fi.mean(axis=1)
    fi_global = fi.sort_values("Global_Mean", ascending=False).head(15)

    print("\n🏁 Top 15 Global RF Features:")
    print(fi_global.to_string())

    print("\n🏁 Top 10 RF Features by Outcome:")
    top_k = 10
    for out in outcomes:
        print("\n" + "="*60)
        print("Outcome:", out)
        print(fi[out].sort_values(ascending=False).head(top_k).to_string())

except Exception as e:
    print("\n FAILED. Exact error:")
    print(type(e).__name__ + ":", str(e))
    traceback.print_exc()
```
